Si el número aleatorio en menor a 0.5 entonces el paso es hacia la izquierda, si es mayor o igual a 0.5 será hacia la izquierda.

```python
pasosDerecha = 0
pasosIzquierda = 0
pasos = 10000
x0 = 0
posiciones = [x0]
for paso in range(pasos):
    decision = np.random.rand()
    if decision < .5:
        pasosIzquierda += 1
        x0 -= 1
        posiciones.append(x0)
    else:
        pasosDerecha += 1
        x0 += 1
        posiciones.append(x0)
```

# Actividad

10 borrachos!

```python
def borracho(pasos):
    posicion_inicial = 0
    x0 = 0
    posiciones = [x0]
    for paso in range(pasos):
        decision = np.random.rand()
        if decision < .5:
            x0 -= 1
            posiciones.append(x0)
        else:
            x0 += 1
            posiciones.append(x0)
    return posiciones
```

```python
for r in range(5):
    plt.plot(borracho(1000), lw = .3, label = "Borracho %s" % r)
    plt.xlabel("Pasos")
    plt.ylabel("Posición")
    plt.legend()
plt.show;
```
